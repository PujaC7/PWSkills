## Q4.  LookupError Exception is the Base class for errors raised when something can't be found. The base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid: IndexError, KeyError.
### An IndexError is raised when a sequence reference is out of range. An IndexError means that your code is trying to access an index that is invalid. This is usually because the index goes out of bounds by being too large. For example, if we have a list with three items and we try to access the fourth item, we will get an IndexError
### the KeyError is raised when accessing an invalid key within a dict. The Python KeyError is a type of LookupError exception and denotes that there was an issue retrieving the key you were looking for. When we see a KeyError , the semantic meaning is that the key being looked for could not be found.
#### example

```python
vehicle = {
  "model": "R15",
  "color": "Black",
}

print(vehicle["speed"])
```
