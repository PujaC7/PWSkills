## Q2.  A scatter plot can suggest various kinds of correlations between variables with a certain confidence interval. For example, weight and height would be on the y-axis, and height would be on the x-axis. Correlations may be positive (rising), negative (falling), or null (uncorrelated).

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("this is my graph")
```

## Q3.  Matplotlib'spyplot API has a convenience function called subplots() which acts as a utility wrapper and helps in creating common layouts of subplots, including the enclosing figure object, in a single call. The two integer arguments to this function specify the number of rows and columns of the subplot grid.

```python
import numpy as np
import matplotlib.pyplot as plt

#line 1:
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 3, 1)
plt.plot(x,y)

#line 2:
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 3, 2)
plt.plot(x,y)

#line 3:
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 3, 3)
plt.plot(x,y)

#line 4:
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 3, 4)
plt.plot(x,y)


plt.show()
```

## Q4.  matplotlib. pyplot. bar is a Python function that creates a vertical bar chart to display data. This function is part of the matplotlib library and is typically used to visualize categorical data.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
  
fig = plt.figure(figsize = (3,3))

plt.bar(company, profit, color ='maroon')

plt.show()
```

## Q5.  The box plot in matplotlib is mainly used to displays a summary of a set of data having properties like minimum, first quartile, median, third quartile, and maximum. The Box Plot is also known as Whisker Plot.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

box = [box1, box2]
 
fig = plt.figure(figsize =(5, 5))
 
ax = fig.add_axes([0, 0, 1, 1])

bp = ax.boxplot(box)

plt.show()
```
